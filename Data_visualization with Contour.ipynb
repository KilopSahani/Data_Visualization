```python

Key Findings:
The most striking observation was the significant increase in rainfall during July over the decades. 
The 90s saw comparatively lower rainfall, while the 21st century witnessed a substantial surge. Contrarily, the months of June, August, 
and September exhibited a gradual decrease in rainfall.

Significance and Alarming Trends:
This revelation, especially the increased rainfall in July, raises alarm bells. The implications of irregular rainfall patterns include severe consequences 
for farmers, leading to yield losses and unfortunately, instances of farmer suicides. 
This alarming trend necessitates immediate attention and proactive solutions to support farmers during these challenging periods.

Call to Action:
Understanding the potential impact on agriculture, we believe in the importance of implementing sustainable solutions for farmers. 
By providing timely information and support, we can empower farmers to navigate these unpredictable weather patterns.
This initiative aims to mitigate the adverse effects on agriculture, ensuring a more stable and secure livelihood for farmers in the North East region.
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import griddata
import matplotlib.lines as mlines

# Use 'YEAR', 'JUN', 'JN-SP' as column names in your DataFrame
X = df['YEAR']
Y = df['JUN']
Z = df['JN-SP']

# Create a grid of points
xi = np.linspace(X.min(), X.max(), 100)
yi = np.linspace(Y.min(), Y.max(), 100)
xi, yi = np.meshgrid(xi, yi)

# Interpolate Z values on this grid
zi = griddata((X, Y), Z, (xi, yi), method='cubic')

# Create the contour plot
plt.contour(xi, yi, zi, levels=14, linewidths=0.5, colors='k')
contour = plt.contourf(xi, yi, zi, levels=14, cmap="RdBu_r")

# Add a colorbar with a label
cbar = plt.colorbar(contour)
cbar.set_label('JN-SP Values')


# Create custom legend
blue_line = mlines.Line2D([], [], markersize=15, label='X: YEAR')
red_line = mlines.Line2D([], [],  markersize=15, label='Y: JUN')
plt.legend(handles=[blue_line, red_line], loc='upper right')


plt.show()
```
